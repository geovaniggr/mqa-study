## História do Dataset

Coletada para Pesquisadora Svetlana Ulianova, da Unviersidade de Ryerson (Canada), conteḿ informações de pacientes em relação a saúde dos pacientes, onde os dados foram coletados no momento da examinação.
O dataset original contem 70.000 instâncias, porém coletamos uma amostra de 5.000, mantendo a precisão de todas as medidas em comparação com os dados originais.


## Variáveis

O dataset contem 12 variáveis, são elas:


**1. Idade (age)**

Quantitativa Discreta

**2. Altura (height)**

Quantitativa Contínua em centrimetros

**3. Peso (weight)**

Quantitativa Contínua em quilos

**4. Genêro (gender)**

Qualitativa Nominal, onde:

    1: Mulher
    2: Homem

**5. Pressão Arterial Sistólica (ap_hi)**

Quantitativa Contínua em milímetros de mercúrio (mmHg)

**6. Pressão Arterial Diastólica (ap_lo)**

Quantitativa Contínua em milímetros de mercúrio (mmHg)

**7. Colesterol (cholesterol)**

Qualitativa Ordinal, onde:

    1: Normal
    2: Acima do Normal
    3: Muito Acima do Normal
    
**8. Glicemia (gluc)**

Qualitativa Ordinal, onde:

    1: Normal
    2: Acima do Normal
    3: Muito Acima do Normal
    
**9. Fumante (smoke)**

Qualitativa Nominal, onde:

    0: Não
    1: Sim

**10. Usuário de Bebidas Alcólicas (alco)**

Qualitativa Nominal, onde:

    1: Não
    0: Sim
    
**11. Pratica Atividade Física? (active)**

Qualitativa Nominal, onde:

    0: Não
    1: Sim

**12. Resultado (cardio)**

Presença ou não de problemas cardiovasculares:

Qualitativa Nominal, onde:

    0: Não
    1: Sim

## Análises Quantitativas (Fumantes)

Questões importantes:

- Dataset Original Não Fumantes: 91.2%

- Dataset Original Fumantes: 8.8%

----

- Amostra Não Fumantes: 91.4%

- Amostra Fumantes: 8.6%

---- 

Média dos não fumantes é cerca de 2 anos mais novo, e não implica uma rotina de vida não saúdavel.

```r
pacientes_fumantes <- database %>% filter(smoke == 1)
pacientes_nao_fumantes <- database %>% filter(smoke == 0)
```

### 1. Fumante x Pressão

```r
boxplot(
    ap_hi ~ smoke,
    xlab = "Tipo de Paciente",
    ylab = "Pressão Sistólica",
    main = "Relação Pressão Sistólica x Tipo de Usuário",
    names = c("Não Fumante", "Fumante"),
    col = c("#3d3c3c", "#d42222"),
    database)
```

```r
summary(pacientes_fumantes$ap_hi)

summary(pacientes_nao_fumantes$ap_hi)
```

```r
boxplot(
    ap_lo ~ smoke,
    xlab = "Tipo de Paciente",
    ylab = "Pressão Sistólica",
    main = "Relação Pressão Sistólica x Tipo de Usuário",
    names = c("Não Fumante", "Fumante"),
    col = c("#3d3c3c", "#d42222"),
    database)
```

```r
summary(pacientes_fumantes$ap_lo)

summary(pacientes_nao_fumantes$ap_lo)
```

```r
## Fumantes

fumantes_doencas_cardio <- pacientes_fumantes %>% filter(cardio == 1)

fumantes_doencas_sem_cardio <- pacientes_fumantes %>% filter(cardio == 0)


nrow(fumantes_doencas_cardio)
nrow(fumantes_doencas_sem_cardio)

proporcao_fumante_doente <- nrow(fumantes_doencas_cardio) / nrow(pacientes_nao_fumantes)
proporcao_fumante_nao_doente <- nrow(fumantes_doencas_sem_cardio) / nrow(pacientes_nao_fumantes)


barplot(
    height = c(nrow(fumantes_doencas_cardio), nrow(fumantes_doencas_sem_cardio)),
    names = c("Fumantes Com Problemas", "Fumantes sem Problemas"),
    col = c("#3d3c3c", "#d42222"))

proporcao_fumante_doente
proporcao_fumante_nao_doente
```

```r
## Fumantes

nao_fumantes_doencas_cardio <- pacientes_nao_fumantes %>% filter(cardio == 1)

nao_fumantes_doencas_sem_cardio <- pacientes_nao_fumantes %>% filter(cardio == 0)


nrow(nao_fumantes_doencas_cardio)
nrow(nao_fumantes_doencas_sem_cardio)


proporcao_nao_fumante_doente <- nrow(nao_fumantes_doencas_cardio) / nrow(pacientes_nao_fumantes)
proporcao_nao_fumante_nao_doente <- nrow(nao_fumantes_doencas_sem_cardio) / nrow(pacientes_nao_fumantes)


barplot(
    height = c(nrow(nao_fumantes_doencas_cardio), nrow(nao_fumantes_doencas_sem_cardio)),
    names = c("Não Fumantes Com Problemas", "Não Fumantes sem Problemas"),
    col = c("#3d3c3c", "#d42222"))

proporcao_nao_fumante_doente
proporcao_nao_fumante_nao_doente
```
