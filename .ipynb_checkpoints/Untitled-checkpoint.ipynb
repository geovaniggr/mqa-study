# Chest Pain x Target

Dores:

- 0: Angina Tipica
- 1: Angina Átpica
- 2: Não Angina
- 3: Asintomático

Target:

- 0: Não Sofreu Ataque
- 1: Sofreu Ataque Cardiáco


```r

chestpain_normal <- nrow(grouped_dataset %>% filter(cp == 0) %>% filter(target == 0))
chestpain_attacked <- nrow(grouped_dataset %>% filter(cp == 0) %>% filter(target == 1))


barplot(height = c(chestpain_normal, chestpain_attacked),
        col = c("#4a4a4a", "#db3737"),
        main = "Angina x Ataque Cardiáco",
        xlab = "Resultado",
        ylab = "Frequência",
        names = c("Não sofreu", "Sofreu"))
```

# Numero de grande vasos coloridos por Fluroscopia

Grandes vasos é um termo usado para se referir coletivamente aos principais vasos sanguíneos, que transportam sangue do e para o coração, são elas:

- Veia cava
    - Veia cava superior
    - Veia cava inferior
- Artéria pulmonar
    - Tronco pulmonar
    - Artéria pulmonar direita
    - Artéria pulmonar esquerda
- Veias pulmonares
    - Superior direita
    - Inferior direita
    - Superior esquerda
    - Inferior esquerda
- Aorta

A fluoroscopia tem várias indicações. As principal em relação ao coração é **estudar o fluxo de sangue nas arteriografias**
