# Classificação 

- Idade: Quantitativa 
- Sexo: Qualilativa 
- Dor no Peito: Qualitativa
- Taxa Cardiáca em Repouso: Quantitativa Continua
- Colesterol Sérico: Quantitatia Continua
- Resultado do Eletrocardiograma em Repouso: Qualitativa
- Máxima Taxa Cardiáca Encontrad: Quantitativa Continua
- Exercicio Gerou Angina? : Qualitativa
- Depressão ST causada pelo exercício: Quantitativa Continua
- Depressão no ST: Qualitativa
- Numéro de Veias Coloridas por Fluorosocopia: Quantitativa Discreta
- Resultado Thal: Qualitativa 


```r
grouped_dataset <- dataset %>%
    mutate( agegroup = case_when( age >= 34 & age < 39 ~ 'G1', 
                                  age >= 39 & age < 44 ~ 'G2',
                                  age >= 44 & age < 49 ~ 'G3',
                                  age >= 49 & age < 54 ~ 'G4',
                                  age >= 54 & age < 59 ~ 'G5',
                                  age >= 59 & age < 64 ~ 'G6', 
                                  age >= 59 & age < 69 ~ 'G7', 
                                  age >= 69 & age < 74 ~ 'G8', 
                                  age >= 74 & age < 79 ~ 'G9'))
```

```r
boxplot(
    chol ~ agegroup,
    grouped_dataset,
    col = base_red,
    main = "Relação Colesterol x Idade",
    xlab = "Grupo de Idades",
    ylab = "Colesterol")
```

```r
boxplot(trestbps ~ agegroup,
    grouped_dataset,
    col = base_red,
    main = "Pressão Sanguínea em Repouso x Idade",
    xlab = "Grupo de Idades",
    ylab = "Pressão Sanguínea em Repouso (mm Hg)")
```

```r
boxplot(thalach ~ agegroup,
    grouped_dataset,
    col = base_red,
    main = "Máxima Taxa Cardiáca durante o Thalium Stress Test x Idade",
    xlab = "Grupo de Idades",
    ylab = "Máxima Taxa Cardiáca")
```

```r
boxplot(oldpeak ~ agegroup,
    grouped_dataset,
    col = base_red,
    main = "Variação no Segmento ST x Idade",
    xlab = "Grupo de Idades",
    ylab = "Variação no Segmento ST (mm)")
```

# Numero de grande vasos coloridos por Fluroscopia

Grandes vasos é um termo usado para se referir coletivamente aos principais vasos sanguíneos, que transportam sangue do e para o coração, são elas:

- Veia cava
    - Veia cava superior
    - Veia cava inferior
- Artéria pulmonar
    - Tronco pulmonar
    - Artéria pulmonar direita
    - Artéria pulmonar esquerda
- Veias pulmonares
    - Superior direita
    - Inferior direita
    - Superior esquerda
    - Inferior esquerda
- Aorta

A fluoroscopia tem várias indicações. As principal em relação ao coração é **estudar o fluxo de sangue nas arteriografias**

## Comparação entre Pacientes


- Os pacientes que não sofreram ataque cardiáco apresentam valores menores de colesterol
- Em relação ao taxa de batimento em repouso tivemos valores parecidos, entretanto há uma diminuida nos batimentos nos pacientes que sofreram ataque
- O desnível nas ondsa ST induzidas por exercício também indicam uma diminuição quando se trata das pessoas que sofreram ataque cardiáco
- Por fim em relação a taxa cardiaca, a média dos usuários que sofreram ataque cardiaco aumentou mais do que os que não tiveram
